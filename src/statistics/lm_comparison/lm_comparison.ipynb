```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_our_lm=pd.read_csv("rank_lm.csv", names=["rank"])
```

```python
df_gpt_lm=pd.read_csv("rank_gpt2.csv", names=["rank"])
```

```python
df_ppl_our_lm=pd.read_csv("ppl_lm.csv", names=["ppl"])
```

```python
df_ppl_gpt_lm=pd.read_csv("ppl_gpt2.csv", names=["ppl"])
```

```python
log=False
bins=30
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
df_our_lm.hist(column = df_our_lm.columns[0], bins = bins, ax=ax[0], figsize=(20, 18), log=log)
df_ppl_our_lm.hist(column = df_ppl_our_lm.columns[0], bins = bins, ax=ax[1], figsize=(20, 18), log=log)
plt.suptitle("Our LM")
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
df_gpt_lm.hist(column = df_our_lm.columns[0], bins = bins, ax=ax[0], figsize=(20, 18), log=log)
df_ppl_gpt_lm.hist(column = df_ppl_our_lm.columns[0], bins = bins, ax=ax[1], figsize=(20, 18), log=log)
plt.suptitle("GPT2 LM")
plt.show()
```
